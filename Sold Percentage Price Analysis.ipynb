```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame.from_csv("ico_sale_data.csv")
```

```python
plt.figure(figsize=(18,10))

df = df[~df['percent_sold'].str.contains("-") & ~df['raised'].str.contains("-")]
df = df[pd.to_numeric(df['raised']) == pd.to_numeric(df['goal'])]

plt.scatter(df['percent_sold'], 
            df['usd_roi'], 
            c=pd.to_numeric(df['raised']), 
            s=pd.to_numeric(df['raised'])/pd.to_numeric(df['goal'])*40)

plt.title('ICO Roi vs % Tokens Sold\nAssuming ICO sold out')
plt.xlabel('Percentage of Tokens Sold')
plt.ylabel('USD ROI Since Sale')

plt.colorbar()
plt.savefig('ico_roi_vs_tokens_sold.png')
plt.show()
```

```python
plt.figure(figsize=(18,10))

df = df[~df['percent_sold'].str.contains("-") & ~df['raised'].str.contains("-")]
df = df[pd.to_numeric(df['raised']) == pd.to_numeric(df['goal'])]

# 
plt.scatter(pd.to_numeric(df['raised'])/pd.to_numeric(df['percent_sold']),
            df['usd_roi'], 
            c=pd.to_numeric(df['percent_sold']), 
            s=pd.to_numeric(df['raised'])/pd.to_numeric(df['goal'])*40)

plt.title('ICO Roi vs % USD Raise per Percent of Tokens Sold\nAssuming ICO sold out')
plt.xlabel('USD raised / Percentage of Tokens Sold')
plt.ylabel('USD ROI Since Sale')

cbar = plt.colorbar()
cbar.set_label('Percentage of Tokens Sold', rotation=270)

plt.savefig('ico_roi_vs_USD_raised_per_percent_sold.png')
plt.show()
```
